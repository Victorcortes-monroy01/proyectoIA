```python
from ultralytics import YOLO

# Cargar un modelo YOLOv8 preentrenado
model = YOLO('yolov8n.pt')  # Cambia 'yolov8n.pt' por otro modelo si lo prefieres

# Entrenar el modelo
model.train(
    data=f"{dataset.location}/data.yaml",  # Ruta al archivo data.yaml
    epochs=50,                             # Número de épocas
    batch=16,                              # Tamaño del batch
    imgsz=640                              # Tamaño de las imágenes
)
```

```python
results = model.predict(source=f"{dataset.location}/test/images", save=True)

```
